```python
# representation within the corpus

import json

file_path = "/Users/zn1010/Documents/Spring 2025/Taxonomy/taxoadapt/datasets/multi_dim/emnlp_2024/final_taxo_tasks.json"
with open(file_path, "r", encoding="utf-8") as file:
    data = json.load(file)

# Function to collect 'paper_ids' for each 'label'
def collect_paper_ids(obj, label_paper_map=None):
    if label_paper_map is None:
        label_paper_map = {}

    if isinstance(obj, dict) and 'label' in obj:
        label = obj['label']
        paper_ids = obj.get('paper_ids', [])
        label_paper_map[label] = paper_ids

        if 'children' in obj and isinstance(obj['children'], list):
            for child in obj['children']:
                collect_paper_ids(child, label_paper_map)

    return label_paper_map

# Collect 'paper_ids' for each 'label'
label_to_paper_ids = collect_paper_ids(data)
label_to_paper_ids
```

```python
title_abstract = []
with open("../taxoadapt/datasets/multi_dim/emnlp_2024/internal.txt", "r") as file:
    for line in file:
        json_object = json.loads(line.strip())
        title_abstract.append("Title: " + json_object['Title'] + " Abstract:" + json_object['Abstract'])
```

```python
from openai import OpenAI
import numpy as np

client = OpenAI()

# Function to compute embeddings
def get_embedding(text, model="text-embedding-3-small"):
   text = text.replace("\n", " ")
   return client.embeddings.create(input = [text], model=model).data[0].embedding

# Function to compute cosine similarity
def cosine_similarity(vec1, vec2):
    vec1 = np.array(vec1)
    vec2 = np.array(vec2)
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
```

```python
corpus_embedding = []
for i in range(len(title_abstract)):
    corpus_embedding.append(get_embedding(title_abstract[i]))
    print(i)
```

```python
print(len(label_to_paper_ids))
node_name_embedding = []
for k in label_to_paper_ids.keys():
    node_name_embedding.append(get_embedding(k))

print(len(node_name_embedding))
```

```python
count, all_sim = 0, []
for k in label_to_paper_ids.keys():
    sum_sim = 0
    for paper_id in label_to_paper_ids[k]:
        sum_sim += cosine_similarity(corpus_embedding[paper_id], node_name_embedding[count])
    if len(label_to_paper_ids[k]) > 0:
        all_sim.append(sum_sim / len(label_to_paper_ids[k]))
    count += 1
    print(count)
    
```

```python
print(len(all_sim))
print("representation within the corpus score:", sum(all_sim) / len(all_sim))
```

```python
num_nodes, num_correct = len(node_name_embedding), 0
for e in node_name_embedding:
    task_sim = cosine_similarity(dim_embedding[0], e)
    methods_sim = cosine_similarity(dim_embedding[1], e)
    datasets_sim = cosine_similarity(dim_embedding[2], e)
    evaluation_sim = cosine_similarity(dim_embedding[3], e)
    real_world_sim = cosine_similarity(dim_embedding[4], e)
    
    max_sim = max([task_sim, methods_sim, datasets_sim, evaluation_sim, real_world_sim])
    if task_sim == max_sim:
        num_correct += 1

print("alignment percentage for task dimension:", num_correct / num_nodes)
```

```python
# How many papers are represented across all dimensional taxonomies? (This is for task dimension)

def find_threshold_and_split(values):
    # Step 1: Sort the list
    sorted_values = sorted(values)
    
    # Step 2: Find the largest gap
    gaps = [(sorted_values[i + 1] - sorted_values[i]) for i in range(len(sorted_values) - 1)]
    max_gap_index = gaps.index(max(gaps))
    
    # Step 3: Define the threshold as the midpoint of the largest gap
    threshold = (sorted_values[max_gap_index] + sorted_values[max_gap_index + 1]) / 2
    
    # Step 4: Split the list
    smaller_part = [v for v in values if v <= threshold]
    larger_part = [v for v in values if v > threshold]
    
    return threshold, smaller_part, larger_part
```

```python
paper_id_set = set()
for k in label_to_paper_ids.keys():
    for paper_id in label_to_paper_ids[k]:
        paper_id_set.add(paper_id)


correct_count = 0
for paper_id in paper_id_set:
    task_sim = cosine_similarity(dim_embedding[0], corpus_embedding[paper_id])
    methods_sim = cosine_similarity(dim_embedding[1], corpus_embedding[paper_id])
    datasets_sim = cosine_similarity(dim_embedding[2], corpus_embedding[paper_id])
    evaluation_sim = cosine_similarity(dim_embedding[3], corpus_embedding[paper_id])
    real_world_sim = cosine_similarity(dim_embedding[4], corpus_embedding[paper_id])

    temp_list = [task_sim, methods_sim, datasets_sim, evaluation_sim, real_world_sim]
    threshold, smaller_part, larger_part = find_threshold_and_split(temp_list)
    if task_sim in larger_part:
        correct_count += 1

```

```python
print("total paper count for task dimension:", correct_count / len(paper_id_set))
```
